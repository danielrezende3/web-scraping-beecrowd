```python
import pandas as pd

tags_csv= pd.read_csv('tags.csv')
beecrowd_problems_csv= pd.read_csv('beecrowd_problems.csv')

# for each same id in both csvs, merge the tags into the beecrowd_problems_csv
df = pd.merge(beecrowd_problems_csv, tags_csv, on='id_number', how='inner', validate='one_to_one')

# save the merged csv
df.to_csv('beecrowd_problems_with_tags.csv', index=False)
```

```python
df = pd.read_csv("beecrowd_problems_with_tags.csv")
df
```

```python
# Carregar o CSV


# Inicializar um conjunto vazio para armazenar as tags únicas
tags_unicas = set()

# Iterar sobre a coluna 'tag'
for tags_str in df["tag"]:
    # Dividir a string em uma lista de tags e atualizar o conjunto
    tags_unicas.update(tags_str.split())

# convert '-' to '_' in all tags_unicas
tags_unicas = set([tag.replace('-', '_') for tag in tags_unicas])
tags_unicas
```

```python
non_math_subtopics = {
    "algorithms_data_structures": [
        "backtracking",
        "binary_search",
        "brute_force",
        "dfs",
        "dp",
        "dynamic_programming",
        "graph",
        "greedy",
        "knapsack",
        "prefix_sums",
        "range_query",
        "recursion",
        "segment_tree",
        "shortest_path",
        "sieve",
        "sorting",
        "subset_sum",
        "trees",
    ],
    "computational_applied_fields": [
        "computational_geometry",
        "game_theory",
        "physics",
        "projectile",
        "relative_motion",
        "scheduling",
        "simulation",
    ],
    "string_formatting": ["string_formatting", "string_gcd"],
    "other": [
        "lfsr",
        "table_calculation",
        "time_calculation",
        "tournament",
        "money_splitting",
    ],
}

# now remove all non_math_subtopics from tags_unicas
for subtopic in non_math_subtopics.values():
    tags_unicas -= set(subtopic)
print(tags_unicas)

# Count the number of problems for each tag in the df
tag_count = {tag: 0 for tag in tags_unicas}
for tags_str in df["tag"]:
    for tag in tags_str.split():
        tag = tag.replace('-', '_')
        if tag in tag_count:
            tag_count[tag] += 1

# sort the tags by count
tag_count = dict(sorted(tag_count.items(), key=lambda item: item[1], reverse=True))
tag_count
```
